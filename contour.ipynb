```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import os
```

```python
line_img = cv2.imread('lines/0.png', cv2.IMREAD_GRAYSCALE)

plt.imshow(line_img, cmap='gray')
```

```python
inv_line_img = np.bitwise_not(line_img)

plt.imshow(inv_line_img, cmap='gray')
plt.show()
```

```python
img_name = 'lines/0.png'
temp_img = cv2.imread(img_name, cv2.IMREAD_GRAYSCALE)
temp_img = np.bitwise_not(temp_img)

contours, _ = cv2.findContours(temp_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

img_reread = np.zeros(temp_img.shape)
filtered_contours = [contour for contour in contours if cv2.contourArea(contour) >= 50]
cv2.drawContours(img_reread, filtered_contours, -1, (255, 255, 255), 2)

plt.imshow(img_reread, cmap='gray')
plt.show()
```
